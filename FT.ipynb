# Mathematical and Numerical Methods - Fourier transform

Download the file `fourier.dat`. This text file is organised in two columns. Column 0 shows the time $t$ in seconds. Column 1 is a noisy signal $y$ measured as a function of time.

### Assignment A
Read columns 0 and 1 of this file into two `numpy` arrays $t$ and $y$ respectively.

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.fft import rfft, irfft
```

```python
t,y=np.genfromtxt("fourier.dat",skip_header=1,usecols=(0,1),unpack="true",delimiter=" ")
```

### Assignment B
Plot $y$ as a function of $t$ with a line plot.

```python
plt.plot(t,y,label="Data")
plt.xlabel("t")
plt.ylabel("y")
plt.legend()
plt.show()
```

### Assignment C
Calculate the Fourier coefficients with the `numpy.fft.rfft` function.

```python
c1=rfft(y)
print(c1[0])
```

### Assignment D
Re-do the same as Assignment C but with a user-provided discrete Fourier transform.

```python
def drft(y):
    N=len(y)
    c=np.zeros(N//2+1,dtype=complex)
    for i in range(N//2+1):
        for j in range(N):
            c[i]+=y[j]*np.exp(-2j*np.pi*i*j/N)
    return c
c2=drft(y)
print(c2[0])
```

### Assignment E
Plot the absolute values of the Fourier coefficients |$c_k$| as a function of $k$.

```python
N=len(y)
C=abs(c2)
n=np.arange(0,N//2+1,1)
plt.plot(n,C)
plt.xlabel("k")
plt.ylabel("|$c_k$|")
plt.show()
```

### Assignment F
Calculate and print the period associated with the signal.

```python
idx=np.argmax(C[1:-1])
T=float(N)/idx
print("Period: ",T)
```
